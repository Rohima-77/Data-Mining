```python
import pandas as pd

datasets = pd.read_csv(
    "/kaggle/input/behavioral-analytics-and-event-frequency/Behavioral Analytics and Event Frequency.csv"
)

datasets.head()
```

```python
N = len(datasets)# Population
print("Total Datasets populations:", N)
n = 100 # Sample Size
```

```python
# Datasets pattern:
# 1. Series Datasets - Single Dimention Datasets
# 2. DataFrame Datasets -  Multi Dimentional Dataset

simple_random_sampling = datasets.sample(n = n,
                                        random_state = 42)
# Random State - Magic Value
# 1, 42
# Best options = 42
print("Type of the sample is :",type(simple_random_sampling))
#print("Simple Random Sampling:", simple_random_sampling)
```

```python
import numpy as np
def sysSampling(datasets, n=100):
    N = len(datasets)
    k = N//n # // for the getting kth value
    starting = np.random.randint(0, k)
    dataIndex = list(range(starting,starting+ k*n,  k))[:n] # Step Size [ starting+ k*n]
    return datasets.iloc[dataIndex]


sysSampling(datasets)
```
