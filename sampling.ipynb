```python
import pandas as pd

a_g_dataset = pd.read_csv('/kaggle/input/age-gender/age_gender.csv')
a_g_dataset.info()
a_g_dataset.shape
```

```python
#dataset filter
age_data=a_g_dataset.dropna(subset=['age'])
age_data=age_data[age_data["age"]>0]
sample_dataset_for_sampling,_=train_test_split(age_data,test_size=0.8,stratify=age_data['gender'],random_state=42)
print(len(sample_dataset_for_sampling))
```

```python
sns.set(style='whitegrid')#Canvas color style
plt.figure(figsize=(8,4)) #canvas area
sns.histplot(data=sample_dataset_for_sampling,
             x='age',
             hue='gender',
             bins=100,
             kde=True,
             palette='Set2',
             stat='density',
             common_norm=False)
plt.title('Stratify Data Sampling')
plt.xlabel('Age Datasets')
plt.ylabel('Gender Datasets')
plt.legend(title='Gender',
          labels=['Female(0)','Male(1)'])
plt.tight_layout()
plt.show()
```
